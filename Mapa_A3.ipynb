```python
import folium

mapa = folium.Map(location=[-14.2350, -51.9253], zoom_start=5)

lojas = [
    ("Super Lojas SP", -23.5505, -46.6333),
    ("Mega Store RJ", -22.9068, -43.1729),
    ("Central BH", -19.9167, -43.9345),
    ("Sul Comércio", -30.0346, -51.2177),
    ("Lojas Norte", -3.1190, -60.0217),
    ("Centro BR", -15.8267, -47.9218),
    ("Nordeste Shops", -8.0476, -34.8770),
    ("Litoral Sul", -12.9714, -38.5016),
]

for nome, lat, lon in lojas:
    folium.Marker([lat, lon], popup=nome).add_to(mapa)

mapa
```
